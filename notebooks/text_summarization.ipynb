```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

device="cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
model_ckpt="google/bigbird-pegasus-large-arxiv"

tokenizer=AutoTokenizer.from_pretrained(model_ckpt)

model=AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)
```

```python
# arxiv_dataset=load_dataset("ccdv/arxiv-summarization")
# arxiv_dataset
```

```python
ARTICLE_TO_SUMMARIZE = """
Scientific inquiry has long been fascinated by the cognitive abilities of avian species.
Recent research has focused on the problem-solving skills of corvids, particularly their
ability to use tools. In a controlled experiment, New Caledonian crows were presented with a
novel puzzle box requiring a sequence of three distinct actions to retrieve a food reward.
The study observed that a significant majority of the subjects (8 out of 10) were able to
solve the puzzle without prior training. Video analysis revealed that the crows engaged in
exploratory behavior, often manipulating the puzzle components in a non-random fashion before
executing the correct sequence. This suggests a high degree of causal reasoning and behavioral
flexibility. The findings contribute to the growing body of evidence that complex cognition is
not exclusive to primates and that species like the New Caledonian crow possess sophisticated
mental faculties for navigating complex challenges. This has implications for our understanding
of the evolution of intelligence.
"""
```

```python
inputs = tokenizer(ARTICLE_TO_SUMMARIZE, max_length=1024, return_tensors="pt", truncation=True)
inputs=inputs.to(device)
```

```python
summary_ids = model.generate(
    inputs['input_ids'],
    num_beams=4,
    min_length=30,
    max_length=150,
    early_stopping=True
)
```

```python
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```
